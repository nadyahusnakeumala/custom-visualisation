<h2>Soal 1:</h2>

Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
fig, ax = plt.subplots (figsize= (10,7))
ax.plot(x,cos, 'r', label = 'cosx')
ax.plot(x,cos3, 'g--', label = '3 cosx')
ax.plot(x,sin, color = 'blue', linestyle = '-.', label = 'sin (x-90)')
ax.plot(x,sin2, 'y*', label = '2*sinx')

ax.set_xlabel('x-axis label')
ax.set_ylabel('y-axis label')
ax.set_title('Grafik Cos Sin')

ax.legend(loc = 'center')
```

<h2>Soal 2: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(nrows = 1, ncols = 3, sharex = True, sharey = True, figsize = (12,5))
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)

ax[0].plot(x,tan, color = 'b')
ax[1].plot(x,cos, color = 'r')
ax[2].plot(x,sin, color = 'g')

ax[0].set_title ('Grafik tanx')
ax[1].set_title ('Grafik cosx')
ax[2].set_title ('Grafik sinx')
plt.show()
```

<h2>Soal 3: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
fig, ax = plt.subplots (figsize = (10,7), sharex = True, sharey = True)
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)

ax.plot(x,y, label = 'Grafik CosX')
ins1 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])


ax.set_title ('Grafik cosx')
ins1.set_title ('Grafik cosx^2')
ins2.set_title ('Grafik cosx^3')
ins3.set_title ('Grafik cosx^4')
ins4.set_title ('Grafik cosx^5')


ins1.plot(x,y2, label = 'Grafik Cos2X', color ='red')
ins1.plot(x,y)
ins2.plot(x,y,x,y3, label = 'Grafik Cos3X', color ='green')
ins2.plot(x,y)
ins3.plot(x,y,x,y4, label = 'Grafik Cos4X', color ='black')
ins3.plot(x,y)
ins4.plot(x,y,x,y5, label = 'Grafik Cos5X', color ='y')
ins4.plot(x,y)

fig.legend()
plt.show()
```
